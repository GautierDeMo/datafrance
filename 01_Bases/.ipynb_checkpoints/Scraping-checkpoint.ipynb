```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
req = requests.get("http://www.journaldunet.com/management/ville/index/villes?page=708")
```

```python
contenu = req.content
```

```python
soup = bs(contenu, "html.parser")
```

```python
groupeListe = soup.findAll('ul', class_='bloc size1of2 odSquareList')
```
